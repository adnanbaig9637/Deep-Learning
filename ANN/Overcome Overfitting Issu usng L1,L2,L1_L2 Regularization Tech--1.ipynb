```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import sklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.datasets import make_classification

import tensorflow
import keras
from keras.models import Sequential
from keras.optimizers import SGD
from keras.layers import InputLayer, Dense
```

```python
#n_sample -- rows
#n_feature--columns
#n_informative = 2-- both the feature should give information
#n_reduantant=0--0 feature should be not importnat feature
#n_repeated column--duplicate clumn
#class_sep -- >1 --class cluster is far way from each other//<1 --class cluster is close and overlapping
#sum of n_informative,n_redundant,n_repeated should equal to n_feature

X,y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_repeated=0, class_sep=4, random_state=1)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y)
#Class sep far away >1
```

```python
#n_sample -- rows
#n_feature--columns
#n_informative = 2-- both the feature should give information
#n_reduantant=0--0 feature should be not importnat feature
#n_repeated column--duplicate clumn
#class_sep -- >1 --class cluster is far way from each other//<1 --class cluster is close and overlapping
#sum of n_informative,n_redundant,n_repeated should equal to n_feature

X,y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_repeated=0, class_sep=1, random_state=1)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y)
#Overlapping class_sep <1
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1, stratify=y)
X_traincv, X_testcv, y_traincv, y_testcv = train_test_split(X_train, y_train, test_size=0.1, random_state=1, stratify=y_train)
#First splitting into train test
#Second splitting into test validation
```

```python
std = StandardScaler()

X_traincv = std.fit_transform(X_traincv)
X_testcv = std.transform(X_testcv)
X_test = std.transform(X_test)
```

```python
model = Sequential()

model.add(InputLayer(input_shape=(2,)))
model.add(Dense(units=64, activation='sigmoid',use_bias=True))
model.add(Dense(units=64, activation='sigmoid', use_bias=True))
model.add(Dense(units=1, activation='sigmoid', use_bias=True))
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_traincv, y_traincv, batch_size=800, epochs=1000, validation_data=(X_testcv, y_testcv), verbose=1)
```

```python
history = model.fit(X_traincv, y_traincv, batch_size=800, epochs=1000, validation_data=(X_testcv, y_testcv), verbose=1)
```

```python
history.history.keys()
```

```python
plt.plot(range(1,1001),history.history['loss'], label='training_loss')
plt.plot(range(1,1001),history.history['val_loss'], label='val_loss')

plt.legend()
plt.show()
```

**TO overcome overfitting issue we will use Regulization techniques**

```python
from keras.regularizers import L1, L2, L1L2
```

```python
l1 = L1(l1=0.01)
l2 = L2(l2=0.01)
l1_l2 = L1L2(l1=0.01,l2=0.01)
```

```python
#Using L1
model_1 = Sequential()

model_1.add(InputLayer(input_shape=(2,)))
model_1.add(Dense(units=64, activation='sigmoid',use_bias=True, kernel_regularizer=l1))
model_1.add(Dense(units=64, activation='sigmoid',use_bias=True, kernel_regularizer=l1))
model_1.add(Dense(units=1, activation='sigmoid', use_bias=True, kernel_regularizer=l1))


model_1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

history = model_1.fit(X_traincv, y_traincv, batch_size=800, epochs=1000, validation_data=(X_testcv, y_testcv), verbose=1)
```

```python
plt.plot(range(1,1001),history.history['loss'], label='training_loss')
plt.plot(range(1,1001),history.history['val_loss'], label='val_loss')

plt.legend()
plt.show()
```

**Now you can see before and after, overfitting issue is reduced using Regularization Techniques**

```python
#Using L2
model_2 = Sequential()

model_2.add(InputLayer(input_shape=(2,)))
model_2.add(Dense(units=64, activation='sigmoid',use_bias=True, kernel_regularizer=l2))
model_2.add(Dense(units=64, activation='sigmoid',use_bias=True, kernel_regularizer=l2))
model_2.add(Dense(units=1, activation='sigmoid', use_bias=True, kernel_regularizer=l2))


model_2.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

history = model_2.fit(X_traincv, y_traincv, batch_size=800, epochs=1000, validation_data=(X_testcv, y_testcv), verbose=1)
```

```python
plt.plot(range(1,1001),history.history['loss'], label='training_loss')
plt.plot(range(1,1001),history.history['val_loss'], label='val_loss')

plt.legend()
plt.show()
```

```python
#Using L1_L2
model_3 = Sequential()

model_3.add(InputLayer(input_shape=(2,)))
model_3.add(Dense(units=64, activation='sigmoid',use_bias=True, kernel_regularizer=l1_l2))
model_3.add(Dense(units=64, activation='sigmoid',use_bias=True, kernel_regularizer=l1_l2))
model_3.add(Dense(units=1, activation='sigmoid', use_bias=True, kernel_regularizer=l1_l2))


model_3.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

history = model_3.fit(X_traincv, y_traincv, batch_size=800, epochs=1000, validation_data=(X_testcv, y_testcv), verbose=1)
```

```python
plt.plot(range(1,1001),history.history['loss'], label='training_loss')
plt.plot(range(1,1001),history.history['val_loss'], label='val_loss')

plt.legend()
plt.show()
```
